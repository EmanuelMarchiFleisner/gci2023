```python
from statsmodels.stats.outliers_influence import  variance_inflation_factor
```

```python
vif = data [[
     'area',
     'densidade',
     'idh',
     'populacao',
     'rendimento',
]]
vif_data = pd.DataFrame()
vif_data['variavel'] = vif.columns
vif_data['VIF'] = [variance_inflation_factor(vif.values, i) for i in range(len(vif.columns))]
vif_data
```

```python
from statsmodels.stats.multicomp import  pairwise_tukeyhsd
from itertools import  combinations
```

```python
tukey =pairwise_tukeyhsd(
    endog=estados['idh'], groups=estados['regiao_nome'], alpha=0.05
)
df_tukey = pd.DataFrame()
combinacao = combinations(tukey.groupunique, 2)
for grupo in list(combinacao):
    df_tukey = df_tukey.apprend(
        {'grupo1': grupo[0], 'grupo 2': grupo[1]}, ignore_index=true)
    df_tukey['reject'] = tukey.reject
    df_tukey['meandiffs'] = tukey.meandiffs
    df_tukey['pvalues'] = tukey.pvalues
    df_tukey.sumary
```
